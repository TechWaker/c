```python
N = 10000
x = sc.random.normal(0,1,N)

x2 = np.array(range(N))/float(N)
x3 = np.sort(x)
plt.title("Probability Distribution Function")
plt.hist(x)
plt.figure()
plt.title("Cumulative Distribution Function")
plt.plot(x3,x2)
```

```python
N = 10000
x = np.random.rayleigh(10,N)
plt.title("Probability Distribution Function")
plt.hist(x)
x2 = np.array(range(N))/float(N)
x3 = np.sort(x)
plt.figure()
plt.title("Cumulative Distribution Function")
plt.plot(x3,x2)
```

```python
x = np.random.exponential(1.0,10000)

plt.title("Probability Distribution Function")
plt.hist(x,bins = 10)
x2 = np.array(range(N))/float(N)
x3 = np.sort(x)
plt.figure()
plt.title("Cumulative Distribution Function")
plt.plot(x3,x2)
```

```python
def pcm(inp,low,up,level):

    delta = (up - low)/level
    print(delta)
    p = []
    for item in inp:
        i = 0
        while(item >= low + i*delta):
            i = i + 1
        p.append(i)
    return [p,delta]
    
def dem(inp,low,up,level):
     delta = (up - low)/level
     p = []
     for item in inp:
         p.append(item*delta + low)
     return p

L = -10
U = 10
t = np.arange(0,2*np.pi,0.1)
x = 10*np.sin(t)

y,y1 = pcm(x,L,U,8)
z1 = dem(y,L,U,8)
y,y2 = pcm(x,L,U,32)
z2 = dem(y,L,U,32)
y,y3 = pcm(x,L,U,128)
z3 = dem(y,L,U,128)
Y = [y1,y2,y3]

plt.title("Pulse Code Modulation")
plt.plot(t,x)
plt.plot(t,z1)
plt.plot(t,z2)
plt.plot(t,z3)
plt.legend(["sin(t)","N = 8","N = 32","N = 128"])
plt.figure()
plt.title("Error")
plt.xlabel("No of levels")
plt.ylabel("Error")
plt.plot([8,32,128],Y)
```
